```python
plt.scatter( x = smart['Calories'], y = smart['TotalSteps'],linestyle = 'dashed')
plt.xlabel('Calories')
plt.ylabel('Total Steps')
plt.show
```

```python
label = ["Very Active Minutes", "Fairly Active Minutes", 
         "Lightly Active Minutes", "Inactive Minutes"]
counts = smart[["VeryActiveMinutes", "FairlyActiveMinutes", 
               "LightlyActiveMinutes", "SedentaryMinutes"]].mean()
explodes = [0.3,0.2,0.1,0.05]
plt.pie(counts, labels = label,explode = explodes, autopct='%.0f%%')
plt.show
```
